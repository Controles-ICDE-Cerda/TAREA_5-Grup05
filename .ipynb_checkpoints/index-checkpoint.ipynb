Superposición espacial: Necesidad de crear o encontrar algunas geometrias de las que se tiene. Se utiliza un enfoque de la teoria de conjuntos

```python
# Se realiza algunas operaciones geometricas
# Se separa las municipalidades por el centroide, según Norte, Sur, Este y Oeste

# North (N) / Norte
MunicipalitiesN_japan=japan_adm2.cx[:,centroidY:]  # japan_adm2 <> Municipalidades de japon
# Southt (S) / Sur
MunicipalitiesS_japan=japan_adm2.cx[:,:centroidY]
# West (W) / Oeste
MunicipalitiesW_japan=japan_adm2.cx[:centroidX,:]
# East (E) / Este
MunicipalitiesE_japan=japan_adm2.cx[centroidX:,:]
```

```python
# Se plotea:

# Norte y Sur se observa que tienen la misma franja en comun (donde pasa el centroide)
base=MunicipalitiesN_japan.plot(facecolor='orange', edgecolor='black',linewidth=0.2, alpha=0.6)
MunicipalitiesS_japan.plot(facecolor='brown', edgecolor='black',linewidth=0.2,ax=base, alpha=0.4)
```

```python
# Se plotea:

# Este y Oste se observa igualmente que tienen una franja en comun (donde pasa el centroide)
base=MunicipalitiesE_japan.plot(facecolor='orange', edgecolor='black',linewidth=0.2, alpha=0.6)
MunicipalitiesW_japan.plot(facecolor='brown', edgecolor='black',linewidth=0.2,ax=base, alpha=0.4)
```

#### Intersección
Se mantiene lo comun entre los los GeoDF

```python
# Se muestran las municipalidades en común ubicados en la franja Norte-Sur (NS)
MunicipalitiesNS_japan=MunicipalitiesN_japan.overlay(MunicipalitiesS_japan, how="intersection",keep_geom_type=True)
MunicipalitiesNS_japan.plot()
```

```python
# Nota: El procedimiento es similar al emplear un spatial join:
MunicipalitiesNS_japan.sjoin(MunicipalitiesS_japan, how="inner", predicate='contains').plot()   #Da el mismo resultado empleaando otro codigo
```

```python
# Manteniendo la superposición, se muestran las municipalidades en común ubicados en la franja Oeste-Este (WE)
MunicipalitiesWE_japan=MunicipalitiesW_japan.overlay(MunicipalitiesE_japan, how="intersection",keep_geom_type=True)
MunicipalitiesWE_japan.plot(edgecolor='white',linewidth=0.1)
```

#### Unión
Se realiza la unión de los dos GeoDF anteriores. Para ello, primero se observa la información de cada uno de ellos:

```python
# Se verifica las columnas presentes en NS
MunicipalitiesNS_japan.info()
```

```python
# Se verifica las columnas presentes en WE
MunicipalitiesWE_japan.info()
```

La superposición combina las geometrías, pero no los atributos. Por ende, se va a crear y mostrar un subconjunto:

```python
# Se unen las intersecciónes de NS y WE

keep=['ADM0_EN_1','geometry']  #Se mantienen estas columnas
MunicipalitiesNS_japan=MunicipalitiesNS_japan.loc[:,keep]
MunicipalitiesWE_japan=MunicipalitiesWE_japan.loc[:,keep]
```

```python
# Ahora, visualizamos los errores y/o celdas vacias que podrian existir
MunicipalitiesNS_japan.overlay(MunicipalitiesWE_japan,how="union",keep_geom_type=True)
```

```python
# Como se pudo observar anteriormente, las geometrias estan bien. No obstante, los atributos no

# Se importa pandas que servira para la intersección de ambos GeoDF (con Concat)
import pandas as pd

pd.concat([MunicipalitiesNS_japan,MunicipalitiesWE_japan],ignore_index=True)
```

Si se desea conservar los atributos, se deberá agregarlos. Pero si desea combinar los resultados finales, simplemente se debe realizar la superposición

```python
# Se plotea la unión
MunicipalitiesNS_japan.dissolve().overlay(MunicipalitiesWE_japan.dissolve(), how="union",keep_geom_type=True).dissolve().plot()
```

Se crea un objeto para guardar el resultado anterior:

```python
MunicipalitiesMidJapan=MunicipalitiesNS_japan.dissolve().overlay(MunicipalitiesWE_japan.dissolve(), how="union",keep_geom_type=True).dissolve()
MunicipalitiesMidJapan
```

```python
MunicipalitiesMidJapan=MunicipalitiesNS_japan.dissolve().overlay(MunicipalitiesWE_japan.dissolve(), how="union",keep_geom_type=True).dissolve()
MunicipalitiesMidJapan

# Se realiza una limpieza
MunicipalitiesMidJapan['zone']='middles'
MunicipalitiesMidJapan=MunicipalitiesMidJapan.loc[:,['ADM0_EN_1_1','zone','geometry']]
MunicipalitiesMidJapan
```

#### **Diferencia**
Aquí, se conserva lo que pertenece al GeoDF de la izquierda que no está en el GeoDF de la derecha:

```python
# Con los municipios:
# Se plotea el resto de la unión
japan_adm2.overlay(MunicipalitiesMidJapan, how='difference').plot()
```

#### **Symmetric Difference (Diferencia simétrica)**
Esto es lo opuesto a *intersección*, conservas lo que no está en la intersección:


```python
# Se realiza una operación de diferencia simetrica entres los municipios del Norte (N) y Sur (S) de Japon. Se visualiza que no se superponen entre si
MunicipalitiesN_japan.overlay(MunicipalitiesS_japan, how="symmetric_difference",keep_geom_type=False).plot()
```

```python
# Se realiza una operación de diferencia simetrica entres los municipios del Oeste (W) y Este (E) de Japon. Se visualiza que no se superponen entre si
MunicipalitiesW_japan.overlay(MunicipalitiesE_japan, how="symmetric_difference",keep_geom_type=False).plot()
```

#### **Validación de la geometría**
Las geometrías se crean de tal manera que pueden aparecer algunos problemas, especialmente en polígonos (multi).
Comprobemos si nuestros mapas recientes sobre estados y municipios son válidos:

```python
# Para filtrar y mostrar las geometrias no validas del GeoDF
japan_adm2[~japan_adm2.is_valid]

# Puesto que algunos poligonos tienen errores, se muestran los que no son válidos
```

```python
# Se plotea los que son invalidos
japan_adm2[~japan_adm2.is_valid].plot()
```

Dado que es difícil de visualizar qué es lo que está mal es necesario el ver algo de información:

```python
# CUAL ES EL ERROR
# Se importa las funciones "explain_validity" y "make_valid" de "shapely.validation".Se utilizaran para para verificar la validación de las geometrias y corregir las no válidas
from shapely.validation import explain_validity, make_valid

explain_validity(japan_adm2[~japan_adm2.is_valid].geometry)
```

```python
# Ahora, se cuentan los errores
japan_adm2['validity']=[x.split('[')[0] for x in japan_adm2.geometry.apply(lambda x: explain_validity(x))]
japan_adm2['validity'].value_counts()
```

```python
# Se deben resolver los errores y se visualizara, de manera que ya no haya errores para las lecturas

# Resolviendo el problema:
japan_adm2.drop(columns=['validity'],inplace=True)
japan_adm2_valid=japan_adm2.copy()
japan_adm2_valid['geometry'] = [make_valid(row)  if not row.is_valid else row for row in japan_adm2_valid['geometry'] ]

# Se va a visualizar si existe alguno que sea no válido
japan_adm2_valid[~japan_adm2_valid.is_valid]
```

La _solution_ que obtuvimos puede ayudar con algunas técnicas avanzadas, pero también puede ocasionarnos algunos problemas adicionales. Una ves que Geopandas resolvio el inconveniente, ha creado **collections**:

```python
# El 3er tipo es "geometry collection" (entre polígonos, líneas, etc), es util para el manejo de datos pero siguen siendo errores para la lectura de shapefiles
pd.Series([type(x) for x in japan_adm2_valid.geometry]).value_counts()
```

Sin embargo, la presencia de colecciones puede resultar problemática en los shapefiles.
